```
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```
history_1 = pd.read_csv('history_1_300.csv',)
history_1 =history_1.rename(columns={'val_accuracy':'bidirectional'})
history_2 = pd.read_csv('history_2_300.csv')
history_2=history_2.rename(columns={'val_accuracy':'lstm'})
history_3 = pd.read_csv('history_3_300.csv')
history_3=history_3.rename(columns={'val_accuracy':'lstm with batchnormalization and dropout'})
history_4 = pd.read_csv('history_4_700.csv')
history_4=history_4.rename(columns={'val_accuracy':'lstm with attention'})
```

```
comb =pd.concat([history_1['bidirectional'],history_2['lstm'],history_3['lstm with batchnormalization and dropout'],history_4['lstm with attention']],axis=1)
```

```
plt.figure(figsize=[ 10, 10])
plt.plot(comb)
plt.title('Model accuracy with truncation 300')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['bidirectional', 'lstm','lstm with batchnormalization and dropout','lstm with attention'],loc='best')
plt.savefig('truncation 300.png')
plt.show()

```
